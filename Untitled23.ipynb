```python
data=pd.read_csv("/content/drive/MyDrive/Week - 2 - Describing the World with Data/Dataset/medical_students_dataset.csv")
df=pd.DataFrame(data)
df.info()
print("\n")
print(f"Null Values: \n\n{df.isnull().sum()}\n")
print(f"Statistics:\n {df.describe()}\n")
numeric_cols=df.select_dtypes(include="number").columns
categorical_cols=df.select_dtypes(include="object").columns
for col in numeric_cols:
  Q1=df[col].quantile(0.25)
  Q3=df[col].quantile(0.75)
  IQR=Q3-Q1
  lower_bound=Q1-1.5*IQR
  upper_bound=Q3+1.5*IQR
  outliers=(df[col]<lower_bound) | (df[col]>upper_bound)

  if outliers.any():
    df[col]=df[col].fillna(df[col].median())
  else:
    df[col]=df[col].fillna(df[col].mean())
print(f"Number of sum after filling Numerical Columns: {df.isnull().sum().sum()}\n")

for col in categorical_cols:
  df[col]=df[col].fillna(df[col].mode()[0])
print(f"Number of sum after filling Categorical Columns: {df.isnull().sum().sum()}\n")

print(f"Duplicates Values: \n{df.duplicated().sum()}")


```

```python
df = pd.read_csv("/content/drive/MyDrive/Week - 2 - Describing the World with Data (1)/Dataset/Titanic-Dataset.csv")

# 1. Calculate mean survival rate for each gender
mean_survival_by_sex = df.groupby('Sex')['Survived'].mean()
print("Mean Survival Rate by Gender:")
print(mean_survival_by_sex)

# 2. Visualization
plt.figure(figsize=(6,4))
sns.barplot(x=mean_survival_by_sex.index, y=mean_survival_by_sex.values, palette='pastel')
plt.title('Mean Survival Rate by Gender')
plt.ylabel('Mean Survival Rate')
plt.xlabel('Sex')
plt.ylim(0,1)
plt.show()
```

```python
df = pd.read_csv("/content/drive/MyDrive/Week - 2 - Describing the World with Data (1)/Dataset/Titanic-Dataset.csv")

# 1. Calculate mean survival rate by Sex and Embarked
mean_survival = df.groupby(['Embarked', 'Sex'])['Survived'].mean().reset_index()
print(mean_survival)

# 2. Visualization: Grouped bar chart
plt.figure(figsize=(8,5))
sns.barplot(x='Embarked', y='Survived', hue='Sex', data=mean_survival, palette='pastel')
plt.title('Mean Survival Rate by Gender and Port of Embarkation')
plt.ylabel('Mean Survival Rate')
plt.xlabel('Port of Embarkation')
plt.ylim(0,1)
plt.show()
```
